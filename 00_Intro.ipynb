```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0, 100, 1000)
noise = np.random.normal(0, 20, 1000)
y = 0.3 * x + noise

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Scatter plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

```
